## Tally Elements

```python
a = ['a', 'a', 'B', '我', '明', 'a', 'B', '7', '23']

def count_elements(seq) -> dict:
     """Tally elements from `seq`."""
     hist = {}
     for k in seq:
         hist[k] = hist.get(k, 0) + 1
     return hist

counted = count_elements(a)
counted
```

```python
from collections import Counter
recounted = Counter(a)
recounted
```

```python
list(counted.items())
#
#f = lambda kv: kv[1]
#f(list(counted.items())[1])
#list(counted.items())[1]
```

```python
sorted_list = sorted(counted.items(), key=lambda kv: kv[1])

for k, v in sorted_list[::-1]:
    print (k, v)
```
