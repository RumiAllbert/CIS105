## Revised Version

````python
houses = ["Eric's house", "Kenny's house", "Kyle's house", "Stan's house"]

def deliver_presents_recursively(houses):
    if len(houses) <= 0:
        print("There is no one available to deliver")
    elif len(houses) == 1:
        house = houses[0]
        print("Delivering presents to", house)
        
    else:
        mid = len(houses)//2
        first_half = houses[:mid]
        second_half = houses[mid:]
        
        deliver_presents_recursively(first_half)
        deliver_presents_recursively(second_half)
        
deliver_presents_recursively(houses)
print('')
deliver_presents_recursively(["Kenny's house" ])
print('')
deliver_presents_recursively([])
````

# Python Recursion

### What is Recursion? 

In short terms, *Recursion* is the way in which a function calls and defines itself. Essentially, recursion is a way, or method, where the solution to a problem depends on smaller instances of the same problem. The concept of recursion is rather difficult to understand, one may think of it as a multi-layered solution to a problem. 

To use an analogy, one may think of recursion as an onion or a cabbage, where there are countless layers that make up the whole vegetable, and moreover, every layer is needed for the vegetable to 'operate' properly. My favorite way to understand recursion is by taking Philippines Island of **Luzon** as an example. The island of Luzon has a lake. This lake contains an island, and this island contains a lake, a this lake contains yet another island. Though the example might not be quite fitting, it can give you an idea of what recursion is. Lets take a look at a basic example to better understand this concept. 

### Simple Example of Recursion

In the following example provided by *GeeksForGeeks*, when printFun(3) is called from main(), the memory is allocated to printFun(3). Then there is a local variable initialized to 3 and statement 1 to 4 are pushed on the stack. The function first prints ‘3’. In statement 2, printFun(2) is called and a local variable is initialized to 2 and statement 1 to 4 are pushed in the stack. Similarly, printFun(2) calls printFun(1) and printFun(1) calls printFun(0). printFun(0) then returns printFun(1). Then again, printFun(1)returns to printFun(2) and so on. In the output, value from 3 to 1 are printed and then 1 to 3 are printed. Output: 3 2 1 1 2 3

```python
def printFun(test): 
  
    if (test < 1): 
        return
    else: 
      
        print( test,end = " ") 
        printFun(test-1) # statement 2 
        print( test,end = " ") 
        return
      
  
test = 3
printFun(test)
```

### Harder Example

Let's take a look at the following example, where the function `linear_sum()` takes a sequence `S` and returns the sum of the first `n` numbers of the sequence.


```python
def linear_sum(S, n):
  """Return the sum of the first n numbers of sequence S."""
  print ("computing sum of " , S[0:n], "n=", n)
  if n == 1:
    return S[0]
  elif n == 2:
        return S[0] + S[1]
  else:
    return linear_sum(S, n-1) + S[n-1]
linear_sum([1,2,3,4], 4)
```
Output:
```
computing sum of  [1, 2, 3, 4] n= 4
computing sum of  [1, 2, 3] n= 3
computing sum of  [1, 2] n= 2
10
```

In the above example, the function `linear_sum()` recursively calls itself until `n` equals to one. From that point, the function sums the items in the sequence:
```
linear_sum([1,2,3,4], 4)
linear_sum([1,2,3], 3)
linear_sum([1,2], 2)
linear_sum([1], 1)

3
3 + 3
6 + 4
Output: 10
```

### Why you should use recursion

Recursion is a great way to simplify your algorithms. This is apparent when you are working on a very big problem. By use of recursion, you can simplify your problem into smaller sub-problems.  

### Why to avoid recursion

Recursion *can* actually be very inefficient in some cases. In the latter example, the recursive calls are rather simple, and quite efficient. However, when we take a problem such as the Fibonacci Sequence, recursion actually becomes very complex and grows exponentially.
