## Problem 3: Write a function to shift a list to its right by one element. 

Your function cycles the last element to fill in the first position of the list. 

 **Note** You are not allowed to use any list methods,  including slicing ([:2], [::-1] etc.)

### Method 1

````python
def shift(aList):
    for i in range(len(aList)-1, 0,  -1):
        k = aList[i]
        aList[i] = aList[i-1]
        aList[i-1] = k 
    
    print(f"\nAfter shifting: myList = {aList}")
````

### Method 2

````python
from collections import deque
def shift(aList):
    a = deque(aList)
    a.rotate()
    print(f"\nAfter shifting: myList = {a}")
````

````python
# Test
myList = [1,2,3,4,5]
print(f"Before shifting, myList = {myList}" )
# Should print
# After shiting, myList=[5, 1, 2, 3, 4]
shift(myList)
````

## Problem 6

> a) Write a function that merge two pre-sorted lists in ascending order such that the merged list is also sorted in ascending order.
>
>You should not use any list functions such as `append()` , `sorted()`, slicing ( `[::-1]` , `[:3]` etc.) 

````python
def merge(L1, L2, Merged):
    # L1, L2: input list, sorted in ascending order     
    # Merged: list to stored merged items, 
    # len(Mereged)==len(L1) + len(L2)
    i = j = k = 0
    while i < len(L1) and j < len(L2):
        if L1[i] < L2[j]:
            Merged[k] = L1[i]
            i += 1
        else:
            Merged[k] = L2[j]
            j += 1
        k += 1
        
    while i < len(L1):
        Merged[k] = L1[i]
        i += 1
        k += 1
        
    while j < len(L2):
        Merged[k] = L2[j]
        j += 1
        k += 1
````

````python
# Test your function:
L = [2, 5, 7, 12, 19]
R = [3, 4, 8, 9]
MergedList = [None]*9
merge(L, R, MergedList)
print (f"MergedList={MergedList}")
# should print 
# MergedList=[2, 3, 4, 5, 7, 8, 9, 12, 19]
````

> b) Characterize time complexity of your function using big-Oh. Please explain your answer.

The above merging algorithm can be characterized using Big O by $O(n)$. In other words, it is in linear time. Since the merging process goes through every array from beginning to end it is linear.

````python
#Exercise 4

def string_reverse(str1):
    rstr1 = ''
    index = len(str1)
    while index > 0:
        rstr1 += str1[index - 1]
        index = index - 1
    return rstr1
print(string_reverse('1234thatcathat'))
````

````python
#Exercise 8

def uniquelist(list):
    x = []
    for k in list:
        if k not in x:
            x.append(k)
    return x

print(uniquelist([1,2,3,3,3,3,4,5])) 
````
